```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import scipy.stats as st
# if using a Jupyter notebook, include:
%matplotlib inline

OUTPUT_PATH = "/Volumes/externo/assets"

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(-2.5,2.5,0.1)
y = np.arange(-2.5,2.5,0.1)
X,Y = np.meshgrid(x,y)
Z = 3*(1-X)**2* np.exp(-(X**2)-(Y+1)**2)-10*(X/5-X**3-Y**5)*np.exp(-X**2-Y**2)-1/3*np.exp(-(X+1)**2-Y**2);


mycmap = plt.get_cmap('gist_earth')
surf = ax.plot_wireframe(X, Y, Z, cmap=mycmap, alpha=0.5)

ax.grid(False)
ax.set_xlabel('$Genotype_1$')
ax.set_ylabel('$Genotype_2$')
ax.set_zlabel('Fitness')

ax.scatter(0.4, -1.9, -6, color='black', s=50)
ax.text(0.4, -1.9, -5.5, "Global Optimum", zorder=1000)

ax.scatter(-0.75, 2.3, -4, s=50, color='black', zorder=1000)
ax.text(-0.75, 2.3, -3, "Local Optimum", zorder=1000)
fig.savefig("%s/fitness_landscape.pdf" % (OUTPUT_PATH), bbox_inches='tight', pad_inches=0.5) 
plt.show()
```
