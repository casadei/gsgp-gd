```python
import os
import re
import pandas as pd
import numpy as np

path = '/Users/casadei/dev/casadei/gsgp-mo/experiments/data'
```

```python
files = os.listdir(path)
```

```python
def read_all():
    
    datasets = {}
    for file in files:
        current = re.sub('-(train|test)-.*', '', file)
        if not current in datasets:
            datasets[current] = {'train': [], 'test': []}

        mode = 'train' if 'train' in file else 'test'
        content = pd.read_csv(path + '/' + file, header=None)
        datasets[current][mode].append([content.shape[0], content.shape[1] -1])
        
    return datasets
    
def summarize(datasets):
    summary = []
    for dataset in datasets:
        current = [dataset]
        
        for mode in ['train', 'test']:
            if len(datasets[dataset][mode]) > 0:
                instances = np.median(list(zip(*datasets[dataset][mode]))[0])
                features = np.median(list(zip(*datasets[dataset][mode]))[1])
            else:
                instances = 0
                features = 0
            
            current.append(len(datasets[dataset][mode]))
            current.append(instances)
            current.append(features)
            
        summary.append(np.array(current))
    
    return np.array(summary)
```

```python
summary = summarize(read_all())
df = pd.DataFrame(data=summary, columns = ('Dataset', 'Tr Folds', 'Training instances', 'Training features', 'Test Folds', 'Test instances', 'Test features'))
df.sort_values(by=['Dataset'])
```
